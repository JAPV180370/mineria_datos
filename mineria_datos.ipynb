# Minería de Datos

## Parcial 1

### Introducción

#### Conceptos

**Minería de Datos:**

Es el proceso de buscar "tesoros" o patrones ocultos dentro de grandes cantidades de información que ya tenemos guardada.

---

##### **Ciencia de Datos**

Es un campo mucho más amplio que abarca todo el ciclo de vida de la información. No solo se encarga de buscar patrones (minería), sino también de capturar los datos, limpiarlos porque vienen desordenados, analizarlos estadísticamente y, lo más importante, crear modelos para predecir el futuro.
