# Minería de Datos

## Parcial 1

### Introducción

La minería de datos se establece como una disciplina interdisciplinaria que combina estadística, inteligencia artificial y sistemas de bases de datos para extraer información implícita y procesable. Su propósito es la transición del almacenamiento masivo de datos hacia la generación de valor estratégico mediante el descubrimiento de estructuras no triviales.

#### Conceptos

**Proceso KDD**

Metodología sistemática que engloba la selección, limpieza y transformación de datos, seguida de la aplicación de algoritmos y la interpretación de los resultados para obtener conocimiento válido.

---

**Modelado Supervisado**

Enfoque orientado a la clasificación y regresión, donde el algoritmo aprende de datos históricos etiquetados para predecir etiquetas o valores numéricos en nuevas instancias.

---

**Modelado No Supervisado**

Técnica enfocada en el descubrimiento de patrones intrínsecos o agrupamientos (clustering) en datos que carecen de etiquetas previas, facilitando la identificación de estructuras ocultas.

---

**Reglas de Asociación**

Análisis de dependencias que identifica la probabilidad de ocurrencia conjunta de eventos, permitiendo descubrir relaciones de causalidad o correlación en grandes conjuntos transaccionales.

##### **Ciencia de Datos**

La Ciencia de Datos constituye el marco holístico que integra la minería de datos junto con técnicas de visualización, ingeniería de datos y computación científica. Mientras que la minería se enfoca específicamente en la extracción de patrones, la Ciencia de Datos abarca todo el ciclo de vida de la información, desde la arquitectura de ingesta hasta la comunicación de hallazgos para la toma de decisiones complejas y la automatización mediante modelos de aprendizaje automático (Machine Learning).
