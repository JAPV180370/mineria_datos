# **Minería de Datos**

## Parcial 1

#### **Introducción**

La minería de datos se establece como una disciplina interdisciplinaria que combina estadística, inteligencia artificial y sistemas de bases de datos para extraer información implícita y procesable. Su propósito es la transición del almacenamiento masivo de datos hacia la generación de valor estratégico mediante el descubrimiento de estructuras no triviales.

---

#### **Conceptos**

**Proceso KDD**

Metodología sistemática que engloba la selección, limpieza y transformación de datos, seguida de la aplicación de algoritmos y la interpretación de los resultados para obtener conocimiento válido.


**Modelado Supervisado**

Enfoque orientado a la clasificación y regresión, donde el algoritmo aprende de datos históricos etiquetados para predecir etiquetas o valores numéricos en nuevas instancias.


**Modelado No Supervisado**

Técnica enfocada en el descubrimiento de patrones intrínsecos o agrupamientos (clustering) en datos que carecen de etiquetas previas, facilitando la identificación de estructuras ocultas.

**Reglas de Asociación**

Análisis de dependencias que identifica la probabilidad de ocurrencia conjunta de eventos, permitiendo descubrir relaciones de causalidad o correlación en grandes conjuntos transaccionales.

---

#### **Ciencia de Datos**

La Ciencia de Datos constituye el marco holístico que integra la minería de datos junto con técnicas de visualización, ingeniería de datos y computación científica. Mientras que la minería se enfoca específicamente en la extracción de patrones, la Ciencia de Datos abarca todo el ciclo de vida de la información, desde la arquitectura de ingesta hasta la comunicación de hallazgos para la toma de decisiones complejas y la automatización mediante modelos de aprendizaje automático (Machine Learning).

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)

print(promedio)
print(promedio_ponderado)
```

```python
# --- Configuración e importaciones ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
# --- Carga y limpieza ---
df = pd.read_csv('/content/drive/MyDrive/MINERIA_DATOS/raw.csv', encoding='latin1')
df.columns = [c.strip() for c in df.columns]
```

```python
# --- Variables derivadas ---
df['edad'] = pd.to_numeric(df['¿Cuántos años cumplidos tienes?'], errors='coerce')
map_sn = {'Si': True, 'No': False}
df['prog_python'] = df['¿Has realizado programas utilizando Python?'].map(map_sn)
df['usa_colab'] = df['¿Has utilizado Google Colab?'].map(map_sn)
df['cuenta_github'] = df['¿Tienes cuenta de GitHub?'].map(map_sn)
```

```python
# --- Paleta de colores ---
accent  = '#2563eb'  # azul
accent2 = '#16a34a'  # verde
accent3 = '#f59e0b'  # ámbar
neutral = '#6b7280'  # gris
```

```python
# =========================
# 1) Histograma de edades
# =========================
fig, ax = plt.subplots(figsize=(8,5))
bins = range(int(df['edad'].min())-1, int(df['edad'].max())+2)
ax.hist(df['edad'], bins=bins, color=accent, edgecolor='white', alpha=0.85)
mean_age = df['edad'].mean()
med_age = df['edad'].median()
ax.axvline(mean_age, color=accent2, linestyle='--', linewidth=2, label=f"Media: {mean_age:.2f}")
ax.axvline(med_age, color=accent3, linestyle=':', linewidth=2.5, label=f"Mediana: {med_age:.2f}")
ax.set_title('Edades recopiladas')
ax.set_xlabel('Años')
ax.set_ylabel('Alumnos')
ax.legend()
plt.tight_layout()
plt.show()
fig.savefig('01_hist_edades.png', dpi=180)
```

```python
# =======================
# 2) Boxplot de edades
# =======================
fig, ax = plt.subplots(figsize=(6,4))
ax.boxplot(df['edad'], vert=False, patch_artist=True,
           boxprops=dict(facecolor=accent, color=accent),
           medianprops=dict(color='white', linewidth=2),
           whiskerprops=dict(color=accent), capprops=dict(color=accent))
ax.set_title('Caja y bigotes: edad')
ax.set_xlabel('Edad (años)')
plt.tight_layout()
plt.show()
fig.savefig('02_box_edades.png', dpi=180)
```

```python
# =======================================
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2×2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','Sí'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','Sí'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
import numpy as np
import pandas as pd
```

```python
anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.

Utilizar el primer punto y el útlimo para trazar una recta:

$$m = \frac{y_{2} - y_{1}}{x_{2} - x_{1}}$$

$$b = y_{1} - m * x_{1}$$

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos $(y_{pred})$: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error:

$$MSE = \frac {\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n}$$

Dónde:

$MSE$ = error cuadrático medio.

$n$ = número de datos.

$Y_i$ = valores observados.

$\hat{Y_{i}}$ = valores predichos.

```python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

Además del MSE (Error Cuadrático Medio), existen varios indicadores clave para evaluar modelos de regresión lineal, cada uno con un enfoque distinto sobre el rendimiento del modelo:

* **RMSE** (Raíz del Error Cuadrático Medio): Es la raíz cuadrada del MSE. A diferencia del MSE, el RMSE expresa el error en las mismas unidades que la variable objetivo, lo que facilita su interpretación. Es útil para penalizar errores grandes, similar al MSE.
* **MAE** (Error Absoluto Medio): Calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales. Es más robusto ante valores atípicos (outliers) que el MSE, ya que no eleva al cuadrado los errores grandes.
* **$R^{2}$** (Coeficiente de Determinación): Indica qué proporción de la variación de la variable dependiente es explicada por el modelo. Va de 0 a 1; un valor más cercano a 1 indica un mejor ajuste.    
* **$R^{2}$** Ajustado: Similar al **$R^{2}$**, pero ajusta el valor según el número de predictores (variables independientes) en el modelo. Es más preciso que el **$R^{2}$** cuando se añaden variables innecesarias, ya que penaliza la complejidad excesiva.   
* **MAPE** (Error Porcentual Absoluto Medio): Mide el error en términos porcentuales, lo que facilita entender la precisión relativa del modelo (ej. "el modelo se equivoca en un 5% en promedio").

Más ejercicios:  

4. Escribe utilizando el Mark Down de Colab las fórmulas de cada uno de los indicadores mencionados arriba.   
5. Calcula los indicadores para la regresión lineal.

*   **RMSE (Raíz del Error Cuadrático Medio)**:

  $$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n}(Y_i - \hat{Y_i})^2}$$

  Donde:
  *   $n$ = número de observaciones
  *   $Y_i$ = valores observados
  *   $\hat{Y_i}$ = valores predichos

*   **MAE (Error Absoluto Medio)**:

  $$MAE = \frac{1}{n} \sum_{i=1}^{n}|Y_i - \hat{Y_i}|$$

  Donde:
  *   $n$ = número de observaciones
  *   $Y_i$ = valores observados
  *   $\hat{Y_i}$ = valores predichos

*   **$R^{2}$ (Coeficiente de Determinación)**:

  $$R^2 = 1 - \frac{\sum_{i=1}^{n}(Y_i - \hat{Y_i})^2}{\sum_{i=1}^{n}(Y_i - \bar{Y})^2}$$

  Donde:
  *   $Y_i$ = valores observados
  *   $\hat{Y_i}$ = valores predichos
  *   $\bar{Y}$ = media de los valores observados

*   **$R^{2}$ Ajustado (Coeficiente de Determinación Ajustado)**:

  $$R_{adj}^{2} = 1 - (1 - R^2) \frac{n-1}{n-p-1}$$

  Donde:
  *   $R^2$ = Coeficiente de Determinación
  *   $n$ = número de observaciones
  *   $p$ = número de predictores (variables independientes)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])

X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

y_pred_sklearn = m_sklearn * anios_escolaridad + b_sklearn
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn
errores_cuadrados_sklearn = errores_sklearn**2
suma_errores_cuadrados_sklearn = np.sum(errores_cuadrados_sklearn)
mse_sklearn = suma_errores_cuadrados_sklearn / len(anios_escolaridad)

y_true = ingreso_mensual_estimado_mx

y_pred = y_pred_sklearn

rmse_sklearn = np.sqrt(mse_sklearn)

mae_sklearn = mean_absolute_error(y_true, y_pred)

r2_sklearn = r2_score(y_true, y_pred)

n = len(y_true)
p = 1
adjusted_r2_sklearn = 1 - (1 - r2_sklearn) * (n - 1) / (n - p - 1)

print(f"RMSE: {rmse_sklearn:.2f}")
print(f"MAE: {mae_sklearn:.2f}")
print(f"R^2: {r2_sklearn:.2f}")
print(f"R^2 ajustado: {adjusted_r2_sklearn:.2f}")
```

Resumen de uso:   
* Precisión general y sensibilidad a outliers: MAE.        
* Penalizar errores grandes: RMSE o MSE.     
* Explicación del modelo: $R^{2}$ o $R^{2}$ Ajustado.         
* Interpretación porcentual: MAPE.

1.  **¿Qué es la probabilidad condicional?**

La probabilidad condicional mide la probabilidad de que ocurra un evento A, sabiendo que ya ocurrió un evento B.

Se denota así:

$$P(A | B)$$

Y su fórmula es:

$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

donde:

*   $P(A \cap B)$: probabilidad de que ocurran A y B a la vez.
*   $P(B)$: probabilidad de que ocurra B.

⚠️ **Importante**: solo tiene sentido si $P(B) > 0$.

2. **Ejemplo básico**

**Ejemplo:**

En una bolsa hay 5 bolas rojas y 3 verdes. Tomamos una bola al azar.

Sea:

*   A = “la bola es roja”
*   B = “la bola es verde o roja” (o sea, cualquier bola…)

Encuentra $P(A | B)$.

Como B es todo el espacio, entonces $P(A | B) = P(A)$.

3. **Ejemplo más interesante**

**Ejemplo:**

En un salón:

*   12 estudiantes hablan inglés
*   8 hablan francés
*   5 hablan ambos idiomas
*   Hay 20 estudiantes en total

Encuentra:

a) P(inglés | francés)

**Interpretación:**

Entre quienes hablan francés, el 62.5% también habla inglés.

**Análisis y Solución del Ejemplo**

**Datos proporcionados:**

*   Total de estudiantes: 20
*   Estudiantes que hablan inglés (Evento A): 12
*   Estudiantes que hablan francés (Evento B): 8
*   Estudiantes que hablan ambos idiomas (A $\cap$ B): 5

**Objetivo:** Encontrar P(inglés | francés), es decir, la probabilidad de que un estudiante hable inglés dado que habla francés.

**Pasos:**

1.  **Calcular la probabilidad de que un estudiante hable francés (P(B)):**

    $$P(B) = \frac{\text{Número de estudiantes que hablan francés}}{\text{Total de estudiantes}} = \frac{8}{20} = 0.40$$

2.  **Calcular la probabilidad de que un estudiante hable inglés y francés (P(A $\cap$ B)):**

    $$P(A \cap B) = \frac{\text{Número de estudiantes que hablan ambos idiomas}}{\text{Total de estudiantes}} = \frac{5}{20} = 0.25$$

3.  **Aplicar la fórmula de probabilidad condicional:**

    $$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

    $$P(\text{inglés | francés}) = \frac{0.25}{0.40} = 0.625$$

**Interpretación:**

La probabilidad de que un estudiante hable inglés, dado que ya se sabe que habla francés, es de 0.625 o 62.5%. Esto significa que, entre el grupo de estudiantes que hablan francés, el 62.5% de ellos también habla inglés.

#### Ejercicios de probabilidad condicional

La población de adultos en un pequeño pueblo que han satisfecho los requisitos para graduarse en la escuela se deben clasificar de acuerdo con su sexo y a si trabajan o no actualmente:

| | Empleado | Desempleado | Total |
| :--- | :---: | :---: | :---: |
| **Hombre** | 460 | 40 | **500** |
| **Mujer** | 140 | 260 | **400** |
| **Total** | **600** | **300** | **900** |

Se seleccionará al azar a uno de estos individuos para que realice un viaje a través de todo el país, con la intención de promocionar las ventajas que se derivan del establecimiento de nuevas industrias en los pequeños poblados. El interés se centra en los siguientes eventos:

* **$H$**: se escoge un hombre.
* **$E$**: el elegido tiene empleo.

**Calcule:** $P(H | E)$

Para calcular la probabilidad de que se escoja un hombre (H) dado que el elegido tiene empleo (E), utilizamos la fórmula de probabilidad condicional:

$$P(H | E) = \frac{P(H \cap E)}{P(E)}$$

Donde:

*   $P(H \cap E)$ es la probabilidad de que un individuo sea hombre Y esté empleado.
*   $P(E)$ es la probabilidad de que un individuo esté empleado.

**Datos de la tabla:**

*   Total de individuos: 900
*   Hombres empleados: 460
*   Total de empleados: 600

**Cálculos:**

1.  **Probabilidad de ser hombre Y empleado ($P(H \cap E)$):**
    $$P(H \cap E) = \frac{\text{Número de hombres empleados}}{\text{Total de individuos}} = \frac{460}{900}$$

2.  **Probabilidad de estar empleado ($P(E)$):**
    $$P(E) = \frac{\text{Número total de empleados}}{\text{Total de individuos}} = \frac{600}{900}$$

3.  **Probabilidad condicional P(H | E):**
    $$P(H | E) = \frac{\frac{460}{900}}{\frac{600}{900}} = \frac{460}{600} \approx 0.767$$

Esto significa que, si seleccionamos al azar a una persona **empleada**, la probabilidad de que esa persona sea un hombre es de aproximadamente **76.7%**.

| Estado / Hábito | No fumadores | Fumadores moderados | Fumadores empedernidos | **Total** |
| :--- | :---: | :---: | :---: | :---: |
| **Hipertenso** | 21 | 36 | 30 | **87** |

| **Total** | **69** | **62** | **49** | **180** |

Si se selecciona aleatoriamente a uno de estos individuos, calcule:

a) Experimente hipertensión, dado que es un fumador empedernido.

b) Sea un no fumador, dado que no ha presentado problemas de hipertensión.

**Solución**

Para resolver estos problemas de probabilidad condicional, utilizaremos los totales calculados de la tabla:
* Total de fumadores empedernidos ($n(E)$): $30 + 19 = 49$
* Total de no hipertensos ($n(NH)$): $48 + 26 + 19 = 93$

a) Probabilidad de hipertensión dado que es fumador empedernido
Buscamos $P(H | E)$, que es la intersección de ser hipertenso y empedernido entre el total de empedernidos:

$$P(H | E) = \frac{30}{49} \approx 0.6122$$

**Resultado:** Existe un **61.22%** de probabilidad de que la persona sea hipertensa si es un fumador empedernido.

b) Probabilidad de ser no fumador dado que no es hipertenso
Buscamos $P(NF | NH)$, que es la intersección de no fumadores no hipertensos entre el total de personas sin hipertensión:

$$P(NF | NH) = \frac{48}{93} \approx 0.5161$$

**Resultado:** Existe un **51.61%** de probabilidad de que la persona sea no fumadora dado que no presenta hipertensión.
